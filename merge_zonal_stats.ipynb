```python
# messy script to merge all zonal stats
import os
import glob
import re
import pandas as pd
from functools import reduce
```

```python
all_data = []
shape_cols = ['id', 'area', 'perimeter']
stats_cols = ['id', '_max', '_min', '_mean', '_std']
for f in glob.glob('/data/mbp14mtp/raster_layers/training_seg_zonal_stats/*zonal_stats.tsv'):
    if os.path.splitext(f)[0].endswith('shape_zonal_stats'):
        all_data.append(pd.read_csv(f, sep='\t', names=shape_cols, index_col='id'))
    else:
        band = re.search('NDevonDart_(.+?)_zonal_stats.tsv', f).group(1)
        print(band)
        colnames = [stats_cols[0]] + [band + x for x in stats_cols[1:]]
        all_data.append(pd.read_csv(f, sep='\t', names=colnames, index_col='id'))

all_data = reduce(lambda x, y: pd.merge(x, y, left_index=True, right_index=True), all_data)
all_data.head()
```

```python
classifications = pd.read_table('/data/mbp14mtp/raster_layers/training_data/ecognition_segments_training_segments_2_classification.tsv',
                                sep='\t', names=['id', 'broadclass', 'class', 'confidence'], index_col='id')
```

```python
all_data = pd.merge(classifications, all_data, left_index=True, right_index=True)
```

```python
all_data.to_csv('/data/mbp14mtp/raster_layers/training_seg_zonal_stats/zonal_stats_merged.tsv',
                sep='\t', float_format='%.2f', index=False)
```
